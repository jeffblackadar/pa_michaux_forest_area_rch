```python
import subprocess
import time

# example command
# C:\LAStools\bin\las2las -i "E:\a_new_orgs\carleton\hist5706-maryland\lidar_files\20120323_001460.laz" -target_utm 18north -o "E:\a_new_orgs\carleton\hist5706-maryland\lidar_files\reprojected\20120323_001460_r_utm.laz"

input_laz_fp = "/content/drive/MyDrive/crane_pennsylvania/laz/pa_forest_district_01/"
output_laz_utm_fp = "/content/drive/MyDrive/crane_pennsylvania/laz/pa_forest_district_01_utm/"
las2las_fp = "/content/drive/MyDrive/github/LAStools/bin/las2las"

import os
#process .laz files one at a time
for root, dirs, files in os.walk(input_laz_fp, topdown=False):
   for name in files:      
      #print(name)
      name_ext = name[-4:].lower()
      if(name_ext==".laz"):
          laz_in_path = os.path.join(input_laz_fp, name)
          laz_out_path = os.path.join(output_laz_utm_fp, (name[:-4]+"_utm"+name_ext))
          command = las2las_fp + " -i " + laz_in_path + " -target_utm 18north -o " + laz_out_path
          print("!"+command)
          subprocess.call(command, shell=True)
          # just in case, let the sub process run
          time.sleep(1)
```

# Break long list of files into folders of 50 files so that they may be downloaded.

```python
import os
import shutil
# Divide files into groups of 200 so they can be downloaded and processed in managably sized chunks.
file_counter = 0
chunk_amount = 50

laz_utm_fp = '/content/drive/MyDrive/crane_pennsylvania/laz/pa_forest_district_01_utm/'
              
laz_utm_fp_dest = os.path.join(laz_utm_fp,'batch_50/')
print("laz_utm_fp_dest",laz_utm_fp_dest)
for file_name in os.listdir(laz_utm_fp):
    origin_file_path = os.path.join(laz_utm_fp,file_name)
    print("origin_file_path", origin_file_path)
    if(os.path.isfile(origin_file_path)):  
        if(file_counter % chunk_amount == 0):
            start_num = (file_counter//chunk_amount)*chunk_amount
            dir_name = "laz_utm_"+ str(start_num)+"-"+str(start_num+(chunk_amount-1))+"/"
            new_dir_path = os.path.join(laz_utm_fp_dest,dir_name)
            print("new_dir_path",new_dir_path)
            if not os.path.exists(new_dir_path):
                os.mkdir(new_dir_path)

        shutil.copy(origin_file_path, os.path.join(new_dir_path,file_name))
        print(file_counter, origin_file_path)
        file_counter = file_counter + 1            

    
```
